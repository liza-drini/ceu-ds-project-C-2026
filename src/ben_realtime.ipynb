## Cleaning dataset

```python
import pandas as pd
rt_df = pd.read_csv('../datasets/Cat_only_rt.csv')
```

```python
rt_df.head()
```

```python
rt_df.dtypes
```

```python
#to drop: griisStatuts, epbcStatus, speciesID, basisOfRecord, imcraRegion
```

```python
rt_df = rt_df.drop(columns = ['Unnamed: 0', 'griisStatus', 'epbcStatus', 'speciesID', 'basisOfRecord', 'imcraRegion'])
```

```python
rt_df = rt_df[rt_df['year'] > 1959]
```

```python
rt_df
```

```python
rt_df.isna().sum()
rt_df = rt_df.dropna()
```

```python
rt_df
```

```python
rt_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(rt_df['year'], rt_df['occurrenceCount'])
plt.xlabel('Years')
plt.ylabel('Occurence Count')
plt.title('Occurence Count of Cats through the years')
plt.show()
```
